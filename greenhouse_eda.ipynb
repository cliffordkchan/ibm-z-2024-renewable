# Greenhouse data plot

The UK aims to hit reduce carbon emissions by 78% from 1990. Let's see if that's true.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datasets/greenhouse.csv')

# co2_emissions = data.iloc[-1][2:] # i give up
# co2_emissions = co2_emissions.to_frame(name='year')
co2_emissions = [
    605418.2, 612356.7, 595916.9, 581129.2, 575102.4, 
    567203.8, 586706.6, 561966.0, 566848.5, 560260.3, 
    567448.4, 576019.9, 557570.3, 568822.0, 569761.6, 
    566156.9, 563224.9, 554200.4, 539075.1, 488335.1, 
    506105.6, 463277.9, 481211.3, 471217.1, 432218.4, 
    416205.2, 393399.4, 381098.8, 374024.1, 359324.7, 
    321087.7
] # copied from last row of table 8.2 (https://www.kaggle.com/datasets/richienrg/uk-greenhouse-gas-emissions-sic-19902020)
# co2_emissions.head()
```

```python
years = []

for yr in range(1990, 1990+ len(co2_emissions)):
    years.append(yr)
    
print(years)
```

```python
# Plotting
plt.figure(figsize=(10, 6))

co2_goal = co2_emissions[0]*(1-0.78)

x = np.array(years)
y = np.array(co2_emissions)
plt.plot(x, y, linestyle = '-', label='CO2 emissions')
plt.axhline(y=co2_goal, color = 'r', linestyle = '-', label='CO2 goal')
plt.legend(loc='upper right')
plt.title('Total CO2 Emissions (1990-2020)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (in metric tons)')
plt.tight_layout()
plt.show()

print("CO2 emissions goal:", co2_goal)
```
