# Exploratory Data Analysis on Renewable datasets

### Downloading packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Reading data

```python
renewable_share_energy = pd.read_csv('datasets/01 renewable-share-energy.csv')
renewable_consumption = pd.read_csv('datasets/02 modern-renewable-energy-consumption.csv')
modern_renewable_prod = pd.read_csv('datasets/03 modern-renewable-prod.csv')
share_electricity_renewables = pd.read_csv('datasets/04 share-electricity-renewables.csv')
hydropower_consumption = pd.read_csv('datasets/05 hydropower-consumption.csv')
hydro_share_energy = pd.read_csv('datasets/06 hydro-share-energy.csv')
share_electricity_hydro = pd.read_csv('datasets/07 share-electricity-hydro.csv')
wind_generation = pd.read_csv('datasets/08 wind-generation.csv')
cumulative_wind_energy_capacity = pd.read_csv('datasets/09 cumulative-installed-wind-energy-capacity-gigawatts.csv')
wind_share_energy = pd.read_csv('datasets/10 wind-share-energy.csv')
share_electricity_wind = pd.read_csv('datasets/11 share-electricity-wind.csv')
solar_energy_consumption = pd.read_csv('datasets/12 solar-energy-consumption.csv')
installed_solar_PV_capacity = pd.read_csv('datasets/13 installed-solar-PV-capacity.csv')
solar_share_energy = pd.read_csv('datasets/14 solar-share-energy.csv')
share_electricity_solar = pd.read_csv('datasets/15 share-electricity-solar.csv')
biofuel_production = pd.read_csv('datasets/16 biofuel-production.csv')
installed_geothermal_capacity = pd.read_csv('datasets/17 installed-geothermal-capacity.csv')

datasets = [renewable_share_energy, 
            renewable_consumption, 
            modern_renewable_prod,
            share_electricity_renewables,
            hydropower_consumption,
            hydro_share_energy,
            share_electricity_hydro,
            wind_generation,
            cumulative_wind_energy_capacity,
            wind_share_energy,
            share_electricity_wind,
            solar_energy_consumption,
            installed_solar_PV_capacity,
            solar_share_energy,
            share_electricity_solar,
            biofuel_production,
            installed_geothermal_capacity]

print(renewable_share_energy.head())
```

```python
for dataset in datasets:
    print(dataset.head())
    print('\n')
```
